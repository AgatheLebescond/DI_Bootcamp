```python
import matplotlib.pyplot as plt

# Data for the bar graph
product_categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Books', 'Toys']
sales_amounts = [25000, 18000, 30000, 12000, 22000]

# Create a bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(product_categories, sales_amounts, color='skyblue')

# Customize the graph
plt.xlabel('Product Categories', fontsize=12)
plt.ylabel('Total Sales Amount ($)', fontsize=12)
plt.title('Retail Company Sales Data by Product Category', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the graph
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Créer une série de dates
dates = pd.date_range('2023-07-01', periods=10, freq='D')

# Créer un MultiIndex avec 'Country', 'City', et 'Date'
index = pd.MultiIndex.from_product([
    ['Canada'],  # Pays
    ['Toronto', 'Vancouver'],  # Ville
    dates  # Date
], names=['Country', 'City', 'Date'])

# Générer des données de température aléatoires
np.random.seed(0)
temperatures = np.random.randint(25, 36, size=20)  # Températures aléatoires entre 25°C et 35°C

# Créer le DataFrame
df_temperatures = pd.DataFrame({'Temperature': temperatures}, index=index)

# Filtrer pour les jours où la température a dépassé 30°C au Canada
hot_days = df_temperatures[df_temperatures['Temperature'] > 30]

print(hot_days)
```

```python
import pandas as pd

# Création d'un DataFrame d'exemple
data = {
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [45000, 55000, 65000, 38000, 72000, 49000]
}

df = pd.DataFrame(data)

# Création d'un index hiérarchique
df.set_index(['Department', 'Employee'], inplace=True)

# Filtrer les lignes où le salaire est supérieur à 50,000
filtered_df = df[df['Salary'] > 50000]

print(filtered_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for MCU movie durations
data = {
    'Movie': [
        'Iron Man', 'The Incredible Hulk', 'Iron Man 2', 'Thor', 'Captain America: The First Avenger',
        'The Avengers', 'Iron Man 3', 'Thor: The Dark World', 'Captain America: The Winter Soldier',
        'Guardians of the Galaxy', 'Avengers: Age of Ultron', 'Ant-Man', 'Captain America: Civil War',
        'Doctor Strange', 'Guardians of the Galaxy Vol. 2', 'Spider-Man: Homecoming', 'Thor: Ragnarok',
        'Black Panther', 'Avengers: Infinity War', 'Ant-Man and The Wasp', 'Captain Marvel',
        'Avengers: Endgame', 'Spider-Man: Far From Home'
    ],
    'Duration': [
        126, 112, 124, 115, 124, 143, 130, 112, 136, 121, 141, 117, 147, 115, 136, 113, 130, 134, 149, 181, 125, 123, 181
    ]
}

# Create a DataFrame
df_mcu = pd.DataFrame(data)

# Create a distribution plot with KDE overlay
plt.figure(figsize=(12, 6))
sns.histplot(df_mcu['Duration'], bins=10, kde=True, color='skyblue')

# Customize the plot
plt.title('Distribution of MCU Movie Durations', fontsize=16)
plt.xlabel('Duration (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for MCU movie scores
scores_data = {
    'Movie': [
        'Iron Man', 'The Incredible Hulk', 'Iron Man 2', 'Thor', 'Captain America: The First Avenger',
        'The Avengers', 'Iron Man 3', 'Thor: The Dark World', 'Captain America: The Winter Soldier',
        'Guardians of the Galaxy', 'Avengers: Age of Ultron', 'Ant-Man', 'Captain America: Civil War',
        'Doctor Strange', 'Guardians of the Galaxy Vol. 2', 'Spider-Man: Homecoming', 'Thor: Ragnarok',
        'Black Panther', 'Avengers: Infinity War', 'Ant-Man and The Wasp', 'Captain Marvel',
        'Avengers: Endgame', 'Spider-Man: Far From Home'
    ],
    'Tomato_Meter': [
        94, 67, 72, 67, 80, 92, 78, 66, 89, 91, 75, 83, 90, 89, 85, 92, 83, 97, 85, 88, 83, 85, 94
    ],
    'Audience_Score': [
        93, 71, 72, 77, 79, 91, 78, 76, 90, 92, 87, 87, 91, 90, 87, 89, 92, 90, 79, 91, 87, 78, 90
    ]
}

# Create a DataFrame
df_scores = pd.DataFrame(scores_data)

# Melt the DataFrame to long format for Seaborn
df_melted = df_scores.melt(id_vars='Movie', value_vars=['Tomato_Meter', 'Audience_Score'], var_name='Score_Type', value_name='Score')

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_melted, x='Score_Type', y='Score', palette='Set2')

# Customize the plot
plt.title('Box Plot of Tomato Meter and Audience Scores for MCU Movies', fontsize=16)
plt.xlabel('Score Type', fontsize=12)
plt.ylabel('Score', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple de données
data = {
    'Production Budget': ['200000000', '150000000', '300000000', '250000000'],
    'Opening Weekend': ['94000000', '85000000', '120000000', '100000000'],
    'Domestic Box Office': ['370000000', '280000000', '450000000', '390000000'],
    'Worldwide Box Office': ['860000000', '700000000', '1000000000', '950000000'],
    'MCU Phase': ['Phase 1', 'Phase 2', 'Phase 3', 'Phase 3']
}

# Création du DataFrame
df = pd.DataFrame(data)

# Conversion des colonnes financières en numérique
financial_columns = ['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']
for column in financial_columns:
    df[column] = df[column].str.replace(',', '').astype(float)

# Création du pairplot
sns.pairplot(df, hue='MCU Phase', vars=financial_columns)
plt.suptitle('Pair Plot of Financial Data by MCU Phase', y=1.02)
plt.show()
```
