Exercise 1: Bar Graph

Here's how you can create a bar graph representing the sales data of a retail company using Matplotlib:

```python

import matplotlib.pyplot as plt

# Sample data
product_categories = ['Electronics', 'Clothing', 'Furniture', 'Books']
sales_amounts = [250000, 180000, 320000, 150000]

# Create bar graph
plt.figure(figsize=(10, 6))
plt.bar(product_categories, sales_amounts, color='skyblue')

# Customize the plot
plt.xlabel('Product Categories')
plt.ylabel('Total Sales Amounts ($)')
plt.title('Sales Data of a Retail Company')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

Exercise 2: Hierarchical Indexing

For this exercise, we'll create a pandas DataFrame with hierarchical indexing and filter the dataset for days when the temperature exceeded 30°C in cities in Canada.

```python
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=10)
countries = ['Canada', 'USA']
cities = ['Toronto', 'Vancouver', 'New York', 'Los Angeles']
data = {
    'Temperature': np.random.randint(20, 40, size=(len(countries) * len(cities) * len(dates)))
}

# Create hierarchical index
index = pd.MultiIndex.from_product([countries, cities, dates], names=['Country', 'City', 'Date'])

# Create DataFrame
df = pd.DataFrame(data, index=index)

# Filter data for temperatures exceeding 30°C in Canada
filtered_df = df[df['Temperature'] > 30].loc['Canada']

print(filtered_df)
```

Exercise 3: Advanced Filtering with Hierarchical Indices

For this exercise, we'll filter rows in a hierarchical DataFrame based on specific criteria.

```python
# Sample data
data = {
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [45000, 55000, 60000, 50000, 65000, 70000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set hierarchical index
df.set_index(['Department', 'Employee'], inplace=True)

# Filter rows where Salary is above 50,000 within each Department
filtered_df = df[df['Salary'] > 50000]

print(filtered_df)
```

Exercise 4: Visualizing Movie Durations

For this exercise, we'll create a distribution plot to visualize the durations of MCU movies using Seaborn.

```python
import seaborn as sns

# Sample data
data = {
    'Movie': ['Iron Man', 'The Incredible Hulk', 'Iron Man 2', 'Thor', 'Captain America: The First Avenger'],
    'Duration': [126, 112, 124, 115, 124]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Duration'], kde=True, color='skyblue')

# Customize the plot
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of MCU Movie Durations')
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Display the plot
plt.show()
```

Exercise 5: Box Plot of Audience and Tomato Meter Scores

For this exercise, we'll compare the distribution of Tomato Meter scores and Audience scores using a box plot.

```python
# Sample data
data = {
    'Movie': ['Iron Man', 'The Incredible Hulk', 'Iron Man 2', 'Thor', 'Captain America: The First Avenger'],
    'Tomato Meter': [94, 67, 72, 69, 80],
    'Audience Score': [91, 71, 72, 68, 79]
}

# Create DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame
df_melted = df.melt(id_vars=['Movie'], value_vars=['Tomato Meter', 'Audience Score'],
                    var_name='Score Type', value_name='Score')

# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Score Type', y='Score', data=df_melted, palette='Set2')

# Customize the plot
plt.xlabel('Score Type')
plt.ylabel('Score')
plt.title('Box Plot of Tomato Meter and Audience Scores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

Exercise 6: Pair Plot of Financial Data

For this exercise, we'll use Seaborn's pairplot to explore relationships between financial aspects of MCU movies.

```python
# Sample data
data = {
    'Movie': ['Iron Man', 'The Incredible Hulk', 'Iron Man 2', 'Thor', 'Captain America: The First Avenger'],
    'Production Budget': [140, 150, 200, 150, 140],
    'Opening Weekend': [98.6, 55.4, 128.1, 65.7, 65.1],
    'Domestic Box Office': [318.4, 134.8, 312.4, 181.0, 176.7],
    'Worldwide Box Office': [585.2, 264.8, 623.9, 449.3, 370.6],
    'MCU Phase': [1, 1, 1, 1, 1]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert financial data to numeric
df[['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']] = df[
    ['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']].apply(pd.to_numeric)

# Create pair plot
sns.pairplot(df, hue='MCU Phase',
             vars=['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office'])

# Display the plot
plt.show()
```
