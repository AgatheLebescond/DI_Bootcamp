## üåü Exercice 3 : Construire un r√©seau neuronal simple avec TensorFlow/Keras
Cr√©ez un r√©seau neuronal simple avec TensorFlow/Keras pour classer les chiffres manuscrits du jeu de donn√©es MNIST. Le r√©seau doit comporter :

Une couche d'entr√©e.
Une couche cach√©e avec 128 neurones et activation ReLU.
Une couche de sortie avec 10 neurones (pour 10 classes) et activation softmax.
Mesures:

1. Importez TensorFlow et les autres modules n√©cessaires, tels que les calques, les mod√®les et les fonctions utilitaires, depuis tensorflow.keras

2. Chargez le jeu de donn√©es MNIST

3. Normalisez les donn√©es en divisant les valeurs de pixels des images par 255,0 (intervalle [0, 1]).

4. Encodez les √©tiquettes en convertissant les √©tiquettes num√©riques (0 √† 9) en vecteurs encod√©s en un seul clic.

5. Pour construire le mod√®le de r√©seau neuronal, cr√©ez un mod√®le s√©quentiel avec l'architecture suivante :

Une couche d'aplatissement pour convertir chaque image 28x28 en un vecteur 1D.
Une couche cach√©e dense avec 128 neurones et activation ReLU.
Une couche dense de sortie avec 10 neurones et activation softmax (pour les 10 classes).

6. Compilez le mod√®le avec les param√®tres suivants :

Optimiseur : ¬´ adam ¬ª
Fonction de perte : ¬´ categorical_crossentropy ¬ª
M√©triques : ['pr√©cision']

7. Entra√Ænez le mod√®le et √©valuez-le sur l‚Äôensemble de donn√©es de test.

Ensemble de donn√©es : L'ensemble de donn√©es MNIST est inclus dans TensorFlow/Keras et peut √™tre charg√© √† l'aide de :

from tensorflow.keras.datasets import mnist

```python
!pip install tensorflow
```

```python
# Imports
import tensorflow as tf
from tensorflow.keras import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Chargement des donn√©es
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 2. Normalisation des images
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# 3. Encodage one-hot des √©tiquettes
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# 4. Construction du mod√®le s√©quentiel
model = Sequential([
    Input(shape=(28, 28)),           # D√©finition propre de l‚Äôentr√©e
    Flatten(),                       # Aplatissement en vecteur 1D
    Dense(128, activation='relu'),   # Couche cach√©e
    Dense(10, activation='softmax')  # Couche de sortie
])
```

```python
# 5. Compilation du mod√®le
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 6. Entra√Ænement du mod√®le
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

### R√©sultats :

* **Pr√©cision entra√Ænement finale** : 98,62‚ÄØ%
* **Pr√©cision validation finale** : 97,68‚ÄØ%
* **Le mod√®le est stable** et apprend efficacement.



```python
# 7. √âvaluation sur l‚Äôensemble de test
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Pr√©cision sur les donn√©es de test : {test_accuracy:.4f}")
```

### R√©sultats sur les donn√©es de test :

* **Pr√©cision :** 97,08‚ÄØ%
* **Perte :** 0.0987

### Analyse rapide :

* L√©g√®re baisse de performance entre validation (97,68‚ÄØ%) et test (97,08‚ÄØ%), ce qui est **normal**.
* Ton mod√®le est **solide, bien g√©n√©ralis√©**, aucune trace de surapprentissage.



```python
# 8. Visualisation d‚Äôune image test avec son √©tiquette
plt.imshow(x_test[0], cmap="gray")
plt.title("√âtiquette r√©elle : " + str(tf.argmax(y_test[0]).numpy()))
plt.axis("off")
plt.show()
```

```python
# 9. Pr√©diction sur une image et affichage du chiffre pr√©dit
image = x_test[0]
prediction = model.predict(np.expand_dims(image, axis=0))
classe_predite = np.argmax(prediction)

plt.imshow(image, cmap="gray")
plt.title(f"Pr√©diction du mod√®le : {classe_predite}")
plt.axis("off")
plt.show()
```

## üåü Exercice 4 : Calcul de la propagation vers l'avant
Dans cet exercice, vous calculerez manuellement la propagation vers l'avant d'un r√©seau neuronal simple qui pr√©dit les prix des maisons en fonction de :

Superficie en pieds carr√©s (x‚ÇÅ)
Nombre de chambres (x‚ÇÇ)
Nous allons calculer le r√©sultat en utilisant les valeurs donn√©es suivantes :

* Input Values:
    x‚ÇÅ = 2000 (Square Footage)
    x‚ÇÇ = 3 (Number of Bedrooms)
* Initial Weights:
     w‚ÇÅ = 0.5 (Weight for Square Footage)
     w‚ÇÇ = 0.7 (Weight for Bedrooms)
* Bias: b = 50,000
* Activation Function: ReLU (Rectified Linear Unit)
* Calculez la valeur de sortie ¬´ z ¬ª avant l‚Äôactivation.
* Appliquez la fonction ReLU pour calculer la pr√©diction finale.
* Interpr√©tez le r√©sultat : quel est le prix pr√©vu de la maison ?

### Donn√©es fournies :

```python
# Entr√©es
x1 = 2000   # Superficie (sq ft)
x2 = 3      # Nombre de chambres

# Poids
w1 = 0.5    # Poids pour la superficie
w2 = 0.7    # Poids pour les chambres

# Biais
b = 50000

# Fonction d‚Äôactivation : ReLU (max(0, z))
def relu(z):
    return max(0, z)
```

### √âtape 1 : Calcul de la sortie avant activation

```python
z = (x1 * w1) + (x2 * w2) + b
print("Valeur avant activation (z) :", z)
```

### √âtape 2 : Application de la fonction ReLU

```python
y_pred = relu(z)
print("Prix pr√©dit de la maison :", y_pred)
```

### R√©sultat final et interpr√©tation :

Le r√©seau neuronal estime que, pour une maison de **2000 pieds carr√©s** avec **3 chambres**, le **prix pr√©dit est 51‚ÄØ002,10‚ÄØ\$**.

Cela signifie que :

* Chaque pied carr√© contribue **0,5‚ÄØ\$** au prix.
* Chaque chambre ajoute **0,7‚ÄØ\$**.
* Le **biais de 50‚ÄØ000‚ÄØ\$** repr√©sente un prix de base (ex. : terrain, localisation...).
* La fonction ReLU assure que le prix ne peut pas √™tre n√©gatif.

**Conclusion :** le mod√®le pr√©dit un prix r√©aliste en combinant les facteurs avec leurs poids + un prix de base fixe.

## Exercice 5 (facultatif) : Impl√©mentation de la propagation avant et arri√®re en Python
Vous coderez un r√©seau neuronal simple qui effectue une propagation vers l'avant et une r√©tropropagation pour un probl√®me de r√©gression (pr√©diction des r√©sultats d'examen en fonction des heures d'√©tude).

import numpy as np

### Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

### Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

### Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

### Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

### Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

### Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

### Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

### Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)


* Ex√©cutez le code et observez comment les poids et les biais sont mis √† jour.
* Expliquez pourquoi la mise √† jour des poids √† l‚Äôaide de la descente de gradient r√©duit l‚Äôerreur.
* Modifiez les poids initiaux ou le taux d‚Äôapprentissage et voyez comment cela affecte l‚Äôapprentissage.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Donn√©es d'entr√©e
x = np.array([4, 80])        # [heures d'√©tude, score pr√©c√©dent]
y_true = 85                  # R√©sultat r√©el de l'examen
```

```python
# Poids et biais initiaux
w = np.array([0.6, 0.3])
b = 10
```

```python
# Taux d‚Äôapprentissage
learning_rate = 0.01
```

```python
# Fonction de propagation avant
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b
    return z  # Pas de ReLU ici car c'est une r√©gression
```

```python
# Fonction de r√©tropropagation
def backward_propagation(x, y_true, y_pred):
    error = y_pred - y_true
    grad_w = error * x
    grad_b = error
    return grad_w, grad_b
```

```python
# √âtape 1 : pr√©diction initiale
y_pred = forward_propagation(x, w, b)
```

```python
# √âtape 2 : calcul de la perte (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2
```

```python
# √âtape 3 : r√©tropropagation
grad_w, grad_b = backward_propagation(x, y_true, y_pred)
```

```python
# √âtape 4 : mise √† jour des param√®tres
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b
```

```python
# Affichage des r√©sultats
print("Pr√©diction initiale :", y_pred)
print("Perte (MSE) :", loss)
print("Poids mis √† jour :", w_new)
print("Biais mis √† jour :", b_new)
```

### R√©sultats obtenus :

* **Pr√©diction initiale** : `36.4`

* **Valeur r√©elle attendue** : `85`

* **Erreur** = `-48.6` ‚Üí le mod√®le **sous-estime fortement** le score.

* **Perte MSE** : `1180.98` (√©lev√©e, logique vu l‚Äôerreur importante)

* **Mise √† jour des poids** :

  * Poids initiaux : `[0.6, 0.3]`
  * Poids apr√®s update : `[2.544, 39.18]`
    Fortes augmentations pour corriger l‚Äôerreur

* **Biais** : de `10` √† `10.486`

---

### Interpr√©tation :

Le mod√®le a ajust√© **brutalement** les poids √† cause :

* d‚Äôun **√©cart important entre pr√©diction et v√©rit√©**
* d‚Äôun **taux d‚Äôapprentissage constant**

  Cela montre que la descente de gradient essaie d‚Äô**augmenter la pr√©diction** rapidement.

## üåü Exercice 6 (facultatif) : Visualisation des pr√©dictions sur l'ensemble de donn√©es MNIST
Entra√Ænez un r√©seau neuronal simple avec TensorFlow/Keras sur le jeu de donn√©es MNIST. Apr√®s l'entra√Ænement, visualisez certaines des pr√©dictions du mod√®le.

Ensemble de donn√©es : l'ensemble de donn√©es MNIST est inclus dans TensorFlow/Keras.

### Voici les √©tapes pour cet exercice :

0. Importer les biblioth√®ques requises : Importer TensorFlow, NumPy, Matplotlib et les autres modules Keras n√©cessaires.
1. Charger le jeu de donn√©es MNIST
. 2. Normaliser les donn√©es : Mettre √† l‚Äô√©chelle les valeurs de pixels sur la plage [0, 1].
3. Encoder les √©tiquettes √† chaud : Convertir les √©tiquettes de classe (chiffres 0 √† 9) en vecteurs encod√©s √† chaud.
4. Construire le mod√®le : Cr√©er un mod√®le avec une couche d‚Äôaplatissement pour convertir chaque image en vecteur, une couche dense avec 128 neurones et activation ReLU, et une couche dense de sortie avec 10 neurones et activation softmax.
5. Compiler le mod√®le avec les param√®tres suivants :

Optimiseur : ¬´ adam ¬ª
Perte : ¬´ categorical_crossentropy ¬ª
M√©trique : ¬´ pr√©cision ¬ª
6. Ajuster le mod√®le sur les donn√©es d'entra√Ænement
7. Faire des pr√©dictions : Utiliser le mod√®le entra√Æn√© pour pr√©dire les √©tiquettes des images de test
8. Visualiser quelques pr√©dictions : Afficher quelques images de test √† c√¥t√© des √©tiquettes pr√©dites en utilisant Matplotlib

```python
# 0. Imports
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Input
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Charger les donn√©es MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 2. Normaliser les images (valeurs entre 0 et 1)
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# 3. Encodage one-hot des √©tiquettes
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python
# 4. Construction du mod√®le
model = Sequential([
    Input(shape=(28, 28)),       # Entr√©e image 28x28
    Flatten(),                   # Aplatir en vecteur
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
# 5. Compilation du mod√®le
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 6. Entra√Ænement du mod√®le
model.fit(x_train, y_train_cat, epochs=5, batch_size=32, validation_split=0.1)
```

```python
# 7. Pr√©dictions sur les images de test
predictions = model.predict(x_test)
predicted_classes = np.argmax(predictions, axis=1)
```

Le mod√®le a fini de faire des pr√©dictions sur l‚Äôensemble de test (313 batches).

```python
# 8. Visualisation de 9 pr√©dictions al√©atoires
plt.figure(figsize=(10, 10))
for i in range(9):
    index = np.random.randint(0, len(x_test))
    plt.subplot(3, 3, i + 1)
    plt.imshow(x_test[index], cmap='gray')
    plt.title(f"Vrai : {y_test[index]}, Pr√©dit : {predicted_classes[index]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

### **Bonnes pr√©dictions :**

* 8/9 sont **correctes** (`4‚Üí4`, `9‚Üí9`, `8‚Üí8`, `2‚Üí2`, `7‚Üí7`, `2‚Üí2`, `9‚Üí9`, `6‚Üí6`)
* Les chiffres sont bien form√©s ‚Üí le mod√®le les reconna√Æt sans difficult√©.

---

### **Erreur rep√©r√©e :**

* `8 ‚Üí 4` : pr√©diction incorrecte (chiffre mal trac√©, peut ressembler √† un "4")

### **Analyse visuelle de l‚Äôerreur :**

* L‚Äôimage du 8 est tr√®s √©cras√©e, presque ouverte en bas.

### **Conclusion :**
