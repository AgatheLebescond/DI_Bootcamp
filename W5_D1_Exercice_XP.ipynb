1. **Différence principale** : L’apprentissage profond utilise des réseaux de neurones à plusieurs couches (profonds), contrairement à l’apprentissage automatique traditionnel qui repose souvent sur des modèles plus simples.

4. **Défis + solutions** :

   * Besoin en données : utiliser des ensembles augmentés ou générés.

5. **Ingénierie des fonctionnalités** : C’est le processus de création manuelle de caractéristiques utiles. Elle est souvent inutile en deep learning car le réseau les apprend automatiquement.

6. **Rôle des couches cachées** : Elles permettent au réseau d’apprendre des représentations hiérarchiques et abstraites des données.

7. **Fonction d’activation** : Elle introduit de la non-linéarité, permettant au réseau de modéliser des relations complexes.

## 🌟 Exercice 3 : Construire un réseau neuronal simple avec TensorFlow/Keras
Créez un réseau neuronal simple avec TensorFlow/Keras pour classer les chiffres manuscrits du jeu de données MNIST. Le réseau doit comporter :

Une couche d'entrée.
Une couche cachée avec 128 neurones et activation ReLU.
Une couche de sortie avec 10 neurones (pour 10 classes) et activation softmax.
Mesures:

1. Importez TensorFlow et les autres modules nécessaires, tels que les calques, les modèles et les fonctions utilitaires, depuis tensorflow.keras

2. Chargez le jeu de données MNIST

3. Normalisez les données en divisant les valeurs de pixels des images par 255,0 (intervalle [0, 1]).

4. Encodez les étiquettes en convertissant les étiquettes numériques (0 à 9) en vecteurs encodés en un seul clic.

5. Pour construire le modèle de réseau neuronal, créez un modèle séquentiel avec l'architecture suivante :

Une couche d'aplatissement pour convertir chaque image 28x28 en un vecteur 1D.
Une couche cachée dense avec 128 neurones et activation ReLU.
Une couche dense de sortie avec 10 neurones et activation softmax (pour les 10 classes).

6. Compilez le modèle avec les paramètres suivants :

Optimiseur : « adam »
Fonction de perte : « categorical_crossentropy »
Métriques : ['précision']

7. Entraînez le modèle et évaluez-le sur l’ensemble de données de test.

Ensemble de données : L'ensemble de données MNIST est inclus dans TensorFlow/Keras et peut être chargé à l'aide de :

from tensorflow.keras.datasets import mnist

```python
# Imports
import tensorflow as tf
from tensorflow.keras import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Chargement des données
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 2. Normalisation des images
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# 3. Encodage one-hot des étiquettes
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# 4. Construction du modèle séquentiel
model = Sequential([
    Input(shape=(28, 28)),           # Définition propre de l’entrée
    Flatten(),                       # Aplatissement en vecteur 1D
    Dense(128, activation='relu'),   # Couche cachée
    Dense(10, activation='softmax')  # Couche de sortie
])
```

```python
# 5. Compilation du modèle
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 6. Entraînement du modèle
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

### Résultats :

* **Précision entraînement finale** : 98,62 %
* **Précision validation finale** : 97,68 %
* **Le modèle est stable** et apprend efficacement.



```python
# 7. Évaluation sur l’ensemble de test
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Précision sur les données de test : {test_accuracy:.4f}")
```

### Résultats sur les données de test :

* **Précision :** 97,08 %
* **Perte :** 0.0987

### Analyse rapide :

* Légère baisse de performance entre validation (97,68 %) et test (97,08 %), ce qui est **normal**.
* Ton modèle est **solide, bien généralisé**, aucune trace de surapprentissage.



```python
# 8. Visualisation d’une image test avec son étiquette
plt.imshow(x_test[0], cmap="gray")
plt.title("Étiquette réelle : " + str(tf.argmax(y_test[0]).numpy()))
plt.axis("off")
plt.show()
```

```python
# 9. Prédiction sur une image et affichage du chiffre prédit
image = x_test[0]
prediction = model.predict(np.expand_dims(image, axis=0))
classe_predite = np.argmax(prediction)

plt.imshow(image, cmap="gray")
plt.title(f"Prédiction du modèle : {classe_predite}")
plt.axis("off")
plt.show()
```

Le modèle a correctement prédit le chiffre : 7.

## 🌟 Exercice 4 : Calcul de la propagation vers l'avant
Dans cet exercice, vous calculerez manuellement la propagation vers l'avant d'un réseau neuronal simple qui prédit les prix des maisons en fonction de :

Superficie en pieds carrés (x₁)
Nombre de chambres (x₂)
Nous allons calculer le résultat en utilisant les valeurs données suivantes :

* Input Values:
    x₁ = 2000 (Square Footage)
    x₂ = 3 (Number of Bedrooms)
* Initial Weights:
     w₁ = 0.5 (Weight for Square Footage)
     w₂ = 0.7 (Weight for Bedrooms)
* Bias: b = 50,000
* Activation Function: ReLU (Rectified Linear Unit)
* Calculez la valeur de sortie « z » avant l’activation.
* Appliquez la fonction ReLU pour calculer la prédiction finale.
* Interprétez le résultat : quel est le prix prévu de la maison ?

### Données fournies :

```python
# Entrées
x1 = 2000   # Superficie (sq ft)
x2 = 3      # Nombre de chambres

# Poids
w1 = 0.5    # Poids pour la superficie
w2 = 0.7    # Poids pour les chambres

# Biais
b = 50000

# Fonction d’activation : ReLU (max(0, z))
def relu(z):
    return max(0, z)
```

### Étape 1 : Calcul de la sortie avant activation

```python
z = (x1 * w1) + (x2 * w2) + b
print("Valeur avant activation (z) :", z)
```

### Étape 2 : Application de la fonction ReLU

```python
y_pred = relu(z)
print("Prix prédit de la maison :", y_pred)
```

### Résultat final et interprétation :

Le réseau neuronal estime que, pour une maison de **2000 pieds carrés** avec **3 chambres**, le **prix prédit est 51 002,10 \$**.

Cela signifie que :

* Chaque pied carré contribue **0,5 \$** au prix.
* Chaque chambre ajoute **0,7 \$**.
* Le **biais de 50 000 \$** représente un prix de base (ex. : terrain, localisation...).
* La fonction ReLU assure que le prix ne peut pas être négatif.

**Conclusion :** le modèle prédit un prix réaliste en combinant les facteurs avec leurs poids + un prix de base fixe.

## Exercice 5 (facultatif) : Implémentation de la propagation avant et arrière en Python
Vous coderez un réseau neuronal simple qui effectue une propagation vers l'avant et une rétropropagation pour un problème de régression (prédiction des résultats d'examen en fonction des heures d'étude).

import numpy as np

### Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

### Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

### Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

### Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

### Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

### Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

### Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

### Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)


* Exécutez le code et observez comment les poids et les biais sont mis à jour.
* Expliquez pourquoi la mise à jour des poids à l’aide de la descente de gradient réduit l’erreur.
* Modifiez les poids initiaux ou le taux d’apprentissage et voyez comment cela affecte l’apprentissage.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Données d'entrée
x = np.array([4, 80])        # [heures d'étude, score précédent]
y_true = 85                  # Résultat réel de l'examen
```

```python
# Poids et biais initiaux
w = np.array([0.6, 0.3])
b = 10
```

```python
# Taux d’apprentissage
learning_rate = 0.01
```

```python
# Fonction de propagation avant
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b
    return z  # Pas de ReLU ici car c'est une régression
```

```python
# Fonction de rétropropagation
def backward_propagation(x, y_true, y_pred):
    error = y_pred - y_true
    grad_w = error * x
    grad_b = error
    return grad_w, grad_b
```

```python
# Étape 1 : prédiction initiale
y_pred = forward_propagation(x, w, b)
```

```python
# Étape 2 : calcul de la perte (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2
```

```python
# Étape 3 : rétropropagation
grad_w, grad_b = backward_propagation(x, y_true, y_pred)
```

```python
# Étape 4 : mise à jour des paramètres
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b
```

```python
# Affichage des résultats
print("Prédiction initiale :", y_pred)
print("Perte (MSE) :", loss)
print("Poids mis à jour :", w_new)
print("Biais mis à jour :", b_new)
```

### Résultats obtenus :

* **Prédiction initiale** : `36.4`

* **Valeur réelle attendue** : `85`

* **Erreur** = `-48.6` → le modèle **sous-estime fortement** le score.

* **Perte MSE** : `1180.98` (élevée, logique vu l’erreur importante)

* **Mise à jour des poids** :

  * Poids initiaux : `[0.6, 0.3]`
  * Poids après update : `[2.544, 39.18]`
    Fortes augmentations pour corriger l’erreur

* **Biais** : de `10` à `10.486`

---

### Interprétation :

Le modèle a ajusté **brutalement** les poids à cause :

* d’un **écart important entre prédiction et vérité**
* d’un **taux d’apprentissage constant**

  Cela montre que la descente de gradient essaie d’**augmenter la prédiction** rapidement.



## Visualisations potentielles

### Visualisation poids avant/après mise à jour

```python
# Comparaison des poids avant/après mise à jour
labels = ['Heures d\'étude', 'Score précédent']
x_axis = np.arange(len(w))

plt.bar(x_axis - 0.15, w, width=0.3, label='Avant')
plt.bar(x_axis + 0.15, w_new, width=0.3, label='Après')
plt.xticks(x_axis, labels)
plt.title("Poids avant et après rétropropagation")
plt.legend()
plt.grid(True)
plt.show()
```

### Interprétation :

```python
# 0. Imports
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Input
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
```
