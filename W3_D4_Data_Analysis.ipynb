```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# Load the dataset
file_path = 'train.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())
```

These rows give us a sense of the variety and range of features present in the dataset, such as battery power, memory, camera specifications, and connectivity options.

```python
# Data Cleaning and Preprocessing
# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())
```

```python
# Statistical Analysis
# Calculate central tendency and variability measures
central_tendency = pd.DataFrame({
    'Mean': data.mean(),
    'Median': data.median(),
    'Mode': data.mode().iloc[0],
    'Variance': data.var(),
    'Std Dev': data.std(),
    'Skewness': data.apply(skew),
    'Kurtosis': data.apply(kurtosis)
})
```

The analysis of missing values in the dataset shows that there are no missing values in any of the features. Each feature has a count of zero missing values, indicating that the dataset is complete. This is great because it means we don't need to perform any data imputation or additional cleaning for missing values.

```python
print("\nStatistical Summary:")
print(central_tendency)
```

Key Observations:

Battery Power: The mean and median are close, suggesting a fairly symmetric distribution. The range is quite broad, from 501 to 1998.

Binary Features (e.g., blue, dual_sim, four_g): These features have values of 0 or 1, with means around 0.5, indicating a roughly even split between the two categories.

Clock Speed: The mean clock speed is 1.52, with a standard deviation of 0.82, indicating some variability but generally moderate clock speeds.

RAM: The mean RAM is 2124.21 with a high standard deviation of 1084.73, indicating a wide range of RAM sizes among the phones.

Price Range: The target variable has a mean of 1.5, indicating an even distribution across the different price categories.

This statistical summary helps us understand the distribution and characteristics of each feature in the dataset.

```python
# Data Visualization
# Set style for the plots
sns.set(style="whitegrid")
```

```python
# Create visualizations for each feature
for feature in data.columns:
    plt.figure(figsize=(14, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data[feature], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

    # Box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=data[feature], color='lightgreen')
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)

    plt.tight_layout()
    plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```


Strong Positive Correlations:

RAM and Price Range: There is a strong positive correlation (0.92), indicating that phones with higher RAM tend to fall into higher price ranges.
Pixel Width and Pixel Height: These features are strongly positively correlated (0.91), which makes sense as they both relate to screen resolution.
Moderate Positive Correlations:

Battery Power and Price Range: There is a moderate positive correlation (0.21), suggesting that higher battery power is somewhat associated with higher price ranges.
Internal Memory and RAM: These features show a moderate positive correlation (0.57), indicating that phones with higher internal memory also tend to have higher RAM.
Strong Negative Correlations:

There are no strong negative correlations evident in this heatmap, which suggests that there are no features that inversely vary strongly with each other in this dataset.
Weak or No Correlations:

Features like Bluetooth (blue), Dual SIM (dual_sim), and Wifi show weak correlations with the price range, indicating that these features do not strongly influence the price range of the phones.
Clock Speed and Mobile Depth (m_dep) also show weak correlations with most other features, suggesting they are relatively independent characteristics in this dataset.
Other Observations:

Screen Height (sc_h) and Screen Width (sc_w) have a moderate positive correlation (0.57), which is expected as they are both related to the physical size of the screen.
Front Camera (fc) and Primary Camera (pc) have a weak positive correlation (0.18), indicating some relationship but not a strong one.
Overall, the heatmap helps identify which features are most strongly related to each other and to the target variable, price range. This information can be useful for feature selection and understanding the key determinants of mobile phone prices

Conclusion

In this analysis, we explored a dataset containing various features of mobile phones and their respective price ranges. Here are the key findings and conclusions:

Key Determinants:

RAM: One of the most significant factors influencing the price range of mobile phones. Phones with higher RAM tend to be in higher price brackets.
Internal Memory: Closely related to RAM, phones with larger internal memory also tend to be priced higher.
Battery Power: While not as influential as RAM or internal memory, higher battery power is somewhat associated with higher price ranges.
Screen Resolution: Features like pixel width and height, which relate to screen resolution, also show a correlation with price, indicating that higher resolution screens are found in more expensive phones.
Unexpected Findings:

Bluetooth and Wi-Fi: These connectivity features show weak correlations with price range, suggesting they are not significant determinants of a phone's price.
Clock Speed and Mobile Depth: These features also show weak correlations with most other features, indicating they are relatively independent and do not strongly influence the price.
Binary Features: Features like dual SIM support, 4G, and 3G capabilities are almost evenly distributed and do not show strong correlations with the price range, indicating that these features are common across various price points.
