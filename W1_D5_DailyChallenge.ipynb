```python
#DailyChallenge Week 1 Day 5, last of this week
# Tasks to do today:
# Be able to add two circles together and return a new circle with the new radius - use a dunder method
# Be able to compare two circles to see which is bigger, and return a Boolean - use a dunder method
# Be able to put them in a list and sort them
# First we need to import math
import math

# A Circle can be defined by either specifying the radius or the diameter.
# The user can query the circle for either its radius or diameter.

#Step 1 Initialization : 
# The goal is to create a class that represents a simple circle.

class Circle:
    def __init__(self, radius=None, diameter=None):
        if radius is not None and diameter is None:
            self.radius = radius
        elif diameter is not None and radius is None:
            self.radius = diameter / 2
        else:
            raise ValueError("You must specify either radius or diameter, but not both.")
        
# Step 2 Diameter Property 

    @property
    def diameter(self):
        return self.radius * 2
 
 # Step 3 Area Property 
 # The area is a calculated property based on the radius. 
 # This makes it easy to get the area of the circle.

    @property
    def area(self):
        return math.pi * self.radius ** 2
    
# Step 4 Representation
# __repr__ provides a clear representation of the object, useful for debugging and display.  
# __str__ does the same here (for simplicity, it’s identical to __repr__)
   
    def __repr__(self):
        return f"Circle(radius={self.radius:.2f})"

    def __str__(self):
        return self.__repr__()
    
# Step 5 Adding Two Circles

    def __add__(self, other):
        if isinstance(other, Circle):
            new_radius = self.radius + other.radius
            return Circle(radius=new_radius)
        raise TypeError("Can only add another Circle")
    
# Step 6 Comparing Size
# Allows comparing the size of two circles (by their radius) using the `<` operator.  
# If you compare with something other than a circle, an error is raised
    def __lt__(self, other):
        if isinstance(other, Circle):
            return self.radius < other.radius
        raise TypeError("Can only compare with another Circle")
 
  
# Step 7 Comparing Equality :
#Allows checking if two circles are equal (same radius).  
#If you compare with something other than a circle, the result is `False`.
    def __eq__(self, other):
        if isinstance(other, Circle):
            return self.radius == other.radius
        return False

#Step 8 Sorting Circles in a List last one for today, et voilà

circles = [Circle(radius=4), Circle(radius=2), Circle(radius=5)]
circles_sorted = sorted(circles)
```
