3. Data Visualization:

```python
import matplotlib.pyplot as plt
import numpy as np

# Générer des données d'exemple
np.random.seed(0)
n_points = 20
cities = ['City 1', 'City 2', 'City 3', 'City 4']
data = {city: np.random.randn(n_points) for city in cities}

# Tracer les lignes
for city, style in zip(cities, ['-', '--', '-.', ':']):
    plt.plot(data[city], label=city, linestyle=style, alpha=0.7)

# Ajouter la légende
plt.legend()

# Afficher le graphique
plt.show()
```
